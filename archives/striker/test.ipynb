```python
from scipy.spatial.transform import Rotation as Rot
import numpy as np
import matplotlib.pyplot as plt
```

```python
ALPHA = np.linspace(0, 1, 10)
```

```python
def plot_tube(ax, a, b, c, r, color = 'gray'):
    v1 = (b - a) / np.linalg.norm(b - a)
    v2 = (b - c) / np.linalg.norm(b - c)
    n = 4
    th = 2 * np.pi * (np.arange(n) / n)
    r1 = Rot.from_rotvec(np.outer(th, v1))
    r2 = Rot.from_rotvec(np.outer(th, v2))
    p1 = np.cross(v1, np.array([1, 0, 0]))
    p1 *= r / np.linalg.norm(p1)
    p2 = np.cross(v2, np.array([1, 0, 0]))
    p2 *= r / np.linalg.norm(p2)
    offsets1 = r1.apply(p1)
    offsets2 = r2.apply(p2)
    for i in range(n):
        line1 = np.outer(ALPHA, a + offsets1[i]) + np.outer(1 - ALPHA, b)
        line2 = np.outer(ALPHA, b) + np.outer(1 - ALPHA, c + offsets2[i])
        ax.plot3D(line1[:, 0], line1[:, 1], line1[:, 2], color)
        ax.plot3D(line2[:, 0], line2[:, 1], line2[:, 2], color)
```

```python
ax = plt.axes(projection='3d')
plot_tube(ax, np.array([0, 0, 1]), np.array([0, 0, 0]), np.array([0, 1, 0]), 0.01)
```
